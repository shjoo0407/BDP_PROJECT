```python
import pandas as pd
```

```python
data2 = pd.read_csv("/Users/janghyun/Desktop/BDP/data/대여이력 정보/data/대여이력 정보2.csv")
data3 = pd.read_csv("/Users/janghyun/Desktop/BDP/data/대여이력 정보/data/대여이력 정보3.csv")
```

```python
data2.columns
```

```python
#data2의 불필요한 column Unamed: 0 제거
data2.drop(['Unnamed: 0'], axis = 1, inplace = True)
data2.columns
```

```python
data3.columns
```

```python
#data2와 data3 column명 통일
data3.columns=['자전거번호', '대여일시', '대여소번호', '대여대여소명', '대여거치대', '반납일시', '반납대여소번호','반납대여소명', '반납거치대', '이용시간', '이용거리']
data3.columns
```

```python
data = pd.concat([data2, data3], ignore_index=True)
print("전체 데이터 개수: ", len(data))

print("데이터 구조: ")
data
```

```python
#날짜 관련 데이터 DateTime형태로 변환
# data['대여일시'] = str(data['대여일시'])
data["date"]=pd.to_datetime(data['대여일시'].str.strip(), format="%Y-%m-%d %H:%M", errors='coerce')
# data['Time'] = pd.to_datetime(df['Time']

# data["대여년도"]= pd.to_datetime(data["대여일시"].str.strip(), dayfirst=True)
# data["대여일시"] = pd.to_datetime(data["대여일시"].str.strip())
# data["반납일시"] = pd.to_datetime(data["반납일시"].str.strip())

# pd.to_datetime(data["대여일시"])
# pd.to_datatime(data["반납일시"])
```

```python
data["대여년도"] = data["date"].dt.year
data.head()
```

```python
#data에 있는 대여년도 파악
data["대여년도"].unique()
```

```python
data2015 = data.groupby("대여년도").get_group(2015.0)
```

```python
#불필요한 column 제거
data2015.drop(["date", "대여년도"], axis = 1, inplace = True)
data2015.head()
```

```python
#data2015 파일로 저장
data2015.to_csv("data2015.csv", encoding="utf-8-sig")
```

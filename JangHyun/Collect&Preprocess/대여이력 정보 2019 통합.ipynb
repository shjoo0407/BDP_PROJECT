```python
import pandas as pd
```

```python
data1 = pd.read_csv("/Users/janghyun/Desktop/BDP/data/대여이력 정보/data2019.csv")
data2 = pd.read_csv("/Users/janghyun/Desktop/BDP/data/대여이력 정보/대여이력 정보 2019.csv")
```

```python
#data1의 불필요한 column Unamed: 0 제거
data1.drop(['Unnamed: 0'], axis = 1, inplace = True)
data1.columns
```

```python
#data2의 불필요한 column Unamed: 0 제거
data2.drop(['Unnamed: 0'], axis = 1, inplace = True)
data2.columns
```

```python
print("data1 데이터 개수: ", len(data1))
print("data2 데이터 개수: ", len(data2))

data = pd.concat([data1, data2], ignore_index=True)
print("전체 데이터 개수: ", len(data))

print("데이터 구조: ")
data
```

```python
#대여일시 기준으로 정렬
data = data.sort_values("대여일시")
```

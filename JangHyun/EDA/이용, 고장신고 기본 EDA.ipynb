### 2019년 데이터로 EDA 진행

```python
data2019 = pd.read_csv("/Users/janghyun/Desktop/BDP/data/대여이력 정보/data/data2019.csv")
broken_bike = pd.read_csv("/Users/janghyun/Desktop/BDP/data/대여이력 정보/data/broken.csv")
```

```python
#column 확인
data2019.columns
```

```python
#불필요한 column 제거
data2019.drop(['Unnamed: 0', 'date'], axis = 1, inplace = True)
data2019.columns
```

```python
#결측치 확인
data2019.isnull().sum()
```

```python
#대여일시 데이터 date type으로 변경
data2019["대여일시"]=pd.to_datetime(data2019['대여일시'].str.strip(), format="%Y-%m-%d %H:%M", errors='coerce')
data2019.head()
```

```python
#대여일시 기준 정렬
data2019 = data2019.sort_values("대여일시")
data2019
```

```python
#2019년 자전거별 총 이용거리
distribute_dis_year = data2019.groupby("자전거번호")["이용거리"].sum().to_frame(name="총 이용거리(M)")
distribute_dis_year.sort_values("총 이용거리(M)")
```

```python
#2019년 자전거별 총 이용시간
distribute_time_year = data2019.groupby("자전거번호")["이용시간"].sum().to_frame(name="총 이용시간(분)")
distribute_time_year.sort_values("총 이용시간(분)")
```

```python
data2019["month"] = data2019["대여일시"].dt.month
```

```python
# 월별 자전거 이용거리 축적구하고 새로운 칼럼 만들기
distribute_dis_month = data2019.groupby("month")["이용거리"].sum().to_frame(name="총 이용거리(M)")
distribute_dis_month.sort_values("총 이용거리(M)")
```

```python
# 월별 자전거 이용시간 축적구하고 새로운 칼럼 만들기
distribute_time_month = data2019.groupby("month")["이용시간"].sum().to_frame(name="총 이용시간(분)")
distribute_time_month.sort_values("총 이용시간(분)")
```

```python
broken_bike["등록일시"]=pd.to_datetime(broken_bike['등록일시'].str.strip(), format="%Y-%m-%d %H:%M", errors='coerce')
broken_bike
```

```python
broken_bike["year"] = broken_bike["등록일시"].dt.year
broken_bike["month"] = broken_bike["등록일시"].dt.month
broken_bike
```

```python
#년도별 고장 신고 데이터 분류
broken_bike_2015 = broken_bike[broken_bike["year"]==2015]
broken_bike_2016 = broken_bike[broken_bike["year"]==2016]
broken_bike_2017 = broken_bike[broken_bike["year"]==2017]
broken_bike_2018 = broken_bike[broken_bike["year"]==2018]
broken_bike_2019 = broken_bike[broken_bike["year"]==2019]
broken_bike_2020 = broken_bike[broken_bike["year"]==2020]
broken_bike_2021 = broken_bike[broken_bike["year"]==2021]
broken_bike_2022 = broken_bike[broken_bike["year"]==2022]

print("2015년 고장신고 건수: ", len(broken_bike_2015))
print("2016년 고장신고 건수: ", len(broken_bike_2016))
print("2017년 고장신고 건수: ", len(broken_bike_2017))
print("2018년 고장신고 건수: ", len(broken_bike_2018))
print("2019년 고장신고 건수: ", len(broken_bike_2019))
print("2020년 고장신고 건수: ", len(broken_bike_2020))
print("2021년 고장신고 건수: ", len(broken_bike_2021))
print("202년 고장신고 건수: ", len(broken_bike_2022))
print("전체 고장신고 건수: ", len(broken_bike))
```
